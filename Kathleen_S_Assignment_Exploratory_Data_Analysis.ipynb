** Create a new dataframe called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

** We can use pandas pct_change() method on the Close column to create a new dataframe representing this return value. Use .groupby().**

```python
returns = append_data.groupby('Company').Close.pct_change()
append_data['Returns'] = returns
append_data
```

** Using this returns DataFrame, figure out on what dates each bank stock had the best and worst single day returns. Did anything significant happen that day?**

```python
# your code here
maxreturn = append_data[append_data.groupby('Company').Returns.transform(max)== append_data["Returns"]]
minreturn = append_data[append_data.groupby('Company').Returns.transform(min)== append_data["Returns"]]
maxmin = pd.concat([maxreturn,minreturn])
maxmin
```

```python
import pandas as pd
import sqlite3
```

```python
cb = sqlite3.connect("crunchbase_analysis.db")
```

```python
#Convert Cruncbase Companies Table Into Dataframe
crunchbase_companies_df = pd.read_sql_query("SELECT * FROM crunchbase_companies;", cb)
crunchbase_companies_df
```

```python
#Clean Crunchbase Companies DF, removing any rows with "None" or "NaN" values
companies_df = crunchbase_companies_df.dropna(axis = 0, how ='any') 
```

```python
#Comparing Dataframe Size
print("Old data frame length:", len(crunchbase_companies_df), "\nNew data frame length:",  
       len(companies_df), "\nNumber of rows with at least 1 NA value: ", 
       (len(crunchbase_companies_df)-len(companies_df))) 
```

```python
companies_df
```

```python
#Convert Cruncbase Investments Table Into Dataframe
crunchbase_investments_df = pd.read_sql_query("SELECT * FROM crunchbase_investments;", cb)
crunchbase_investments_df
```

```python
#Clean Crunchbase Investments DF, removing any rows with "None" or "NaN" values
investments_df = crunchbase_investments_df.dropna(axis = 0, how ='any') 
investments_df 
```

```python
#Convert Cruncbase Acquisitions Table Into Dataframe
crunchbase_acquisitions_df = pd.read_sql_query("SELECT * FROM crunchbase_acquisitions;", cb)
crunchbase_acquisitions_df
```

```python
#Clean Crunchbase Acquisitions DF, removing any rows with "None" or "NaN" values
acquisitions_df = crunchbase_acquisitions_df.dropna(axis = 0, how ='any') 
acquisitions_df 
```

```python
#Full join Companies and Investments Table, and remove rows with missing matches
df1 = pd.merge(companies_df, investments_df, left_on='permalink', right_on='company_permalink', how='outer')
df1 = df1.dropna(axis = 0, how ='any') 
df1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
#Exploratory pairplot graph of companies and investments dataframe, color coded by status
sns.set(style="ticks")
sns.pairplot(df1, hue="status")
```

```python
#Count of company listing by category code
sns.countplot(df1['category_code'],palette="plasma")
fig = plt.gcf()
plt.title('Company Listing by Category Code',fontsize='25')
plt.xlabel("Category Code")
plt.ylabel("Count of Category Code")
plt.xticks(rotation=90)
fig.set_size_inches(15,10)
#Count plot illustrates that the top 3 company categories listed in the Crunchbase database are within the software,enterprise, and mobile category.
```

```python
#Count of company listing by funding year
sns.countplot(df1['funded_year'],palette="plasma")
fig = plt.gcf()
plt.title('Companies Funded by Year',fontsize='25')
plt.xlabel("Funding Year")
plt.ylabel("Count of Companies Funded")
plt.xticks(rotation=90)
fig.set_size_inches(15,10)
#Count plot illustrates that the top 3 years where the most companies were funded, occurred in 2011-2013. Specifically, looking at 2013, this was the year that today's modern startups began to boom.
```

```python
#Count of company listing by company region
sns.countplot(df1['region'],palette="plasma")
fig = plt.gcf()
plt.title('Company Listing by Region',fontsize='25')
plt.xlabel("Company Region")
plt.ylabel("Count of Company Region")
plt.xticks(rotation=90)
fig.set_size_inches(15,10)
#The top 3 regions for company listings in the database are SF Bay, NYC, and Boston
```

```python
#Let's focus on NY region and look the top 10 company categories invested in by investors in this region.
sf_investor_companies = df1.loc[df1['region'] == 'New York']
fig2 = sf_investor_companies.groupby('category_code').name.count().plot(kind='bar', figsize=(15,10), color="blue")
plt.title('Company Listing by Category in NY Region',fontsize='25')
plt.xlabel("Category Code")
plt.ylabel("Number of Companies")
#If we focus on just NY region, investors are mostly investing in web, software, and advertising companies.
```

```python
#Let's look at the funding round type for these companies in NYC
sf_investor_companies = df1.loc[df1['region'] == 'New York']
fig2 = sf_investor_companies.groupby('funding_round_type').name.count().plot(kind='bar', figsize=(15,10), color="pink")
plt.title('Funding Round Type for Companies in NY Region',fontsize='25')
plt.xlabel("Funding Round Type")
plt.ylabel("Number of Companies")
#Here, we find that most companies in NY region are currently in the angel and series a funding.
```
